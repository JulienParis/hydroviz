```python
### -- pesticides / MCT (moy concentrations totales)  --
lab_MCT = "MCT"

def setMLIndex (df, year):
    row_labels_mct = list(df['CD_STATION'])
    index = pd.MultiIndex.from_product([row_labels_mct,year], names=['CD_STATION', 'ANNEE'])
    #print index
    return index

def multilevel_MCT(df, year) :

    # delete column year with inplace
    df.drop("ANNEE", axis=1, inplace=True)
    
    # add column CD_PARAMETRE, LB_PARAMETRE
    df["CD_PARAMETRE"] = 9999
    df["LB_PARAMETRE"] = "all pesticides"
    
    # set indexes : STATIONS
    df.set_index(["CD_STATION"], inplace=True) 
    #df.set_index(["CD_STATION", "ANNEE"], inplace=True) 
    #df.set_index(["CD_STATION", "CD_PARAMETRE", "LB_PARAMETRE"], inplace=True) 
    
    # get columns labels 
    row_labels_mct = list(df.index.values)
    #print " -- col_labels :", col_labels_mct
    
    # get columns labels 
    col_labels_mct = list(df.columns.values)
    #print " -- col_labels :", col_labels_mct
    
    index = pd.MultiIndex.from_product([row_labels_mct,year], names=['CD_STATION', 'ANNEE'])
    df.index = index

    # add multilevel hierarchy on columns
    #df.columns = pd.MultiIndex.from_product([lab_MCT, col_labels_mct, year])
    #df.columns = pd.MultiIndex.from_product([lab_MCT, col_labels_mct])

    #dataframe.head(8) 
    return df
```

```python
### -- pesticides / MA (moy analyses)  --
lab_MA = "MA"

def multilevel_MA (dataframe, year):

    # set indexes : STATION and CODE_PESTICIDE
    dataframe.set_index(["CD_STATION", "CD_PARAMETRE", "LB_PARAMETRE"], inplace=True)
    #dataframe.set_index(["CD_STATION"], inplace=True)

    # get columns labels for df_ma
    col_labels_df_ma = list(dataframe.columns.values)
    #print " -- col_labels :", col_labels_df_ma

    # add multilevel hierarchy on columns
    dataframe.columns = pd.MultiIndex.from_product([lab_MA, col_labels_df_ma, year])
    
    return dataframe
```

```python
df_ma_2007 = pd.read_csv("datas/"+datas_MA[0], sep=";", encoding = csv_encoding)
df_ma_2007 = multilevel_MA(df_ma_2007, 2007)
    
df_ma_2008 = pd.read_csv("datas/"+datas_MA[1], sep=";", encoding = csv_encoding)
df_ma_2008 = multilevel_MA(df_ma_2008, 2008)

df_ma_2009 = pd.read_csv("datas/"+datas_MA[2], sep=";", encoding = csv_encoding)
df_ma_2009 = multilevel_MA(df_ma_2009, 2009)

df_ma_2010 = pd.read_csv("datas/"+datas_MA[3], sep=";", encoding = csv_encoding)
df_ma_2010 = multilevel_MA(df_ma_2010, 2010)

df_ma_2011 = pd.read_csv("datas/"+datas_MA[4], sep=";", encoding = csv_encoding)
df_ma_2011 = multilevel_MA(df_ma_2011, 2011)

df_ma_2012 = pd.read_csv("datas/"+datas_MA[5], sep=";", encoding = csv_encoding)
df_ma_2012 = multilevel_MA(df_ma_2012, 2012)
```
